```python
# prompt: cambiar nombres de columnas al español

import pandas as pd

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Cambiar nombres de columnas al español
nombres_columnas_español = {
    'sepal length (cm)': 'longitud_sepalo_cm',
    'sepal width (cm)': 'ancho_sepalo_cm',
    'petal length (cm)': 'longitud_petalo_cm',
    'petal width (cm)': 'ancho_petalo_cm'
}
iris_df = iris_df.rename(columns=nombres_columnas_español)

# Ahora puedes usar iris_df con los nombres de columnas en español
print(iris_df.head())

fig,ax = plt.subplots(figsize=(7,5))
ax.scatter(iris_df['longitud_sepalo_cm'], iris_df['ancho_sepalo_cm'], c=iris.target)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.plotting.scatter_matrix(iris_df, c=iris.target, figsize=(10, 10), s=60, alpha=.8) # Use iris_df instead of df
plt.show()
```

# OneHotEncoder - Modifica Variables Categoricas a Numericas

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

adult_df = pd.read_csv("/content/adult.csv")
adult_df.head()

print (adult_df.sex.unique())

```

```python
one_hot_encoder = OneHotEncoder()
encoded_adult_sex = one_hot_encoder.fit_transform(adult_df[['sex']])
print(encoded_adult_sex.toarray())
print(one_hot_encoder.categories_)

adul_df[one_hot_encoder.categories_[0]] = encoded_adult_sex.toarray()
adul_df.head()
```
